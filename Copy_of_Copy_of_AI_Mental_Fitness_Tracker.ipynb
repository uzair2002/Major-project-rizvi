```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import os
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df2= pd.read_csv("mental-and-substance-use-as-share-of-disease -AI.csv")
df1= pd.read_csv("prevalence-by-mental-and-substance-use-disorder _AI.csv")
```

```python
df1.head()

```

```python
df2.head(10)
```

```python
data=pd.merge(df1,df2)
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.drop('Code',axis=1,inplace=True)
```

```python

data.head()
```

```python
data.size,data.shape
```

```python
data.set_axis(['Country','Year','Schizophrenia', 'Bipolar_disorder', 'Eating_disorder','Anxiety','drug_usage','depression','alcohol','mental_fitness'], axis='columns', inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True,cmap='Blues')
plt.plot()
```

```python
sns.jointplot(x='alcohol',y='mental_fitness',data=data,kind='reg',color='c')
plt.show()
```

```python
sns.jointplot(x='depression',y='mental_fitness',data=data,kind='reg',color='orange')
plt.show()
```

```python
sns.jointplot(x='drug_usage',y='mental_fitness',data=data,kind='reg',color='black')
plt.show()
```

```python
sns.jointplot(x='Eating_disorder',y='mental_fitness',data=data,kind='reg',color='g')
plt.show()
```

```python
sns.jointplot(x='Anxiety',y='mental_fitness',data=data,kind='reg',color='r')
plt.show()
```

```python
sns.jointplot(data=data,x='Bipolar_disorder',y='mental_fitness',kind='reg')
plt.show()
```

```python
sns.jointplot(x='Schizophrenia',y='mental_fitness',data=data,kind='reg',color='m')
plt.show()
```
